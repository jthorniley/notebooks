```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
from pathlib import Path
import json
import matplotlib as mpl
from cycler import cycler

def load_colorscheme(name, background="background", foreground="foreground", grid="cyan"):
    definition = json.loads((Path("iTerm2-Color-Schemes") / "windowsterminal" / f"{name}.json").read_bytes())
    mpl.rcParams["axes.edgecolor"] = definition[grid].lstrip("#")
    mpl.rcParams["grid.color"] = definition[grid].lstrip("#")
    mpl.rcParams["figure.facecolor"] = definition[background].lstrip("#")
    mpl.rcParams["axes.facecolor"] = definition[background].lstrip("#")
    mpl.rcParams["legend.facecolor"] = definition[background].lstrip("#")
    mpl.rcParams["legend.edgecolor"] = definition[grid].lstrip("#")
    mpl.rcParams["text.color"] = definition[foreground].lstrip("#")

    colors = (definition[c] for c in [
        "red", "blue", "yellow", "green", "purple", "cyan", "brightRed", "brightGreen", "brightPurple", "brightCyan",
    ] if definition[c] not in (definition[grid], definition[background]))

    mpl.rcParams['axes.prop_cycle'] = cycler(color=colors)
```

```python
from matplotlib.patches import ArrowStyle

load_colorscheme("BlueBerryPie", foreground="red")
plt.figure(figsize=(8,8))
x = np.linspace(0, np.pi)
for i in np.arange(10):
    style = ["-","--","-.",":"][i % 4]
    plt.plot(x, np.sin(x) * (1.0 - i/10), linestyle=style, label=f"Series {i}")
plt.xticks(np.linspace(0, np.pi, 10))
plt.yticks(np.linspace(0, 1, 10))
plt.xlim([-0.1, np.pi+0.1])
plt.ylim([-0.1, 1.1])
plt.legend()
plt.savefig("test.png")
```
